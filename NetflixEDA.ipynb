Data Cleaning:

```python
#find null values in each columns
data.isnull().sum()
```

```python
#find unique values in each column
data.nunique()
```

```python
data["rating"].value_counts()
```

```python
data["rating"].fillna("TV-MA",inplace=True)
```

```python
data["director"].value_counts()
```

```python
data["director"].fillna("unknown",inplace=True)
```

```python
data["cast"].value_counts()
```

```python
data["cast"].fillna("unknown",inplace=True)
```

```python
data["country"].value_counts()
```

```python
data["country"].fillna(data["country"].mode()[0],inplace=True)
```

```python
data.dropna(subset=["date_added"],inplace=True)
```

```python
data.dropna(subset=["duration"],inplace=True)
```

```python
#data shape after removing all the null values
data.shape
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
#change the data type of date_added to datetime
data["date_added"]
data["parsed_date"]=[pd.to_datetime(date,format='%B %d, %Y',errors="coerce") for date in data["date_added"]]
data.drop("date_added",axis=1,inplace=True)

```

```python
data.dropna(subset=["parsed_date"],inplace=True)
```

```python
data.isnull().sum()
```

```python
data["month"]=pd.to_numeric(data["parsed_date"].dt.month).astype("int64")
```

```python
data["year"]=pd.to_numeric(data["parsed_date"].dt.year).astype("int64")
```

```python
data.dtypes
```

```python
data.head(5)
```

```python
data=data.rename(columns={"listed_in":"genre"})
data["genre"]=data["genre"].apply(lambda x: x.split(",")[0])
```
